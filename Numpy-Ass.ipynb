```python
import numpy as np
import pandas as pd
```

```python
#generating the dataset,i will be using 601 because its a range value and the 600 is included.
np.random.seed(42)
days = np. arange(1, 121)
consumption =np.random.randint(200, 601, size=120)


print(consumption)

print("\ntotal days:", days) 
```

```python
data = pd. DataFrame({
    "Days": np.arange(1, 121),
    "Consumption (kwh)": consumption
})
data
```

```python
#basic stats
print("Average consumption:", np.mean(consumption))
print("Median consumption:", np.median(consumption))
print("Minimum consumption:", np.min(consumption))
print("Maximum consumption:", np.max(consumption))
print("consumption variability:", np.std(consumption))

```

```python
#identify extremes
#find the index of max and min consumption
max_day_index = np.argmax(consumption)
min_day_index = np.argmin(consumption)

#convert index to days number by adding 1
day_highest = max_day_index + 1
day_lowest = min_day_index + 1

print("highest consumption:", day_highest)
print("lowest consumption:", day_lowest)
```

```python
#monthly splitting
months = np. array_split (consumption, 4)

print(months)




```

```python
#slicing of the consumption array into four months array.
month1, month2, month3, month4 = np. array_split (consumption, 4)

print("first_month:", month1)
print("second_month:", month2)
print("third_month:", month3)
print("fourth_month:", month4)
```

```python
#6. monthly analysis
weeks = np. array_split (consumption, 17)
print(weeks)

print("week total consumption")
week_total = [print(week.sum()) for week in weeks]

print("\nweek average consumption")
week_avg = [print(week. mean()) for week in weeks]
```

```python
#7. overall consumption trend 
cumulative_consumption = np.cumsum(consumption)

print(cumulative_consumption)

print("First 10 values of Cumulative Consumption (kWh):")

print(cumulative_consumption[:10])
```

```python
#8. anomaly detection
mean_consumption = np.mean(consumption)
std_consumption = np.std(consumption)

lower_bound = mean_consumption - 2* std_consumption
upper_bound = mean_consumption + 2* std_consumption

print(round(lower_bound, 2))
print(round(upper_bound, 2))

outlier = np.where((consumption < lower_bound) | (consumption > upper_bound))[0] + 1

print("Anomaly / outlier detected:", outlier)
```
