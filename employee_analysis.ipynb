```python
#start cleaning inconsistent values
#clean columns name
df.rename(columns={
    "EmpId": "emp_id",
    "fullName": "full_name",
    "Dept": "department",
    "Age": "age",
    "Salary($)": "salary",
    "JoinDate": "join_date",
    "Performance": "performance",
    "Remote?": "remote"
}, inplace=True)

df
```

```python
#clean inconsistence values on full_name


#let check for the values we have on fullname
df['full_name'].to_list()
```

```python
df['full_name'] = (
    df['full_name'] 
    .str.strip()
    .str.title()
    .str.replace(r"[\s-]+", " ", regex=True)
)

df['full_name'].to_list()
```

```python
#clean inconsistence values in department
df.replace({
    "hr": "HR",
    "Mktng": "Marketing",
    "marketing": "Marketing",

}, inplace=True)

df['department']
```

```python
#check the vales of age
df['age'].to_list()
```

```python
#clean inconsistence value in age
def clean_age(value):
   if value == "thirty-five":
       return 35
   elif isinstance(value, (int, float)):
       return value
   elif isinstance (value, str) and value.isdigit():
        return int(value)
   else:
        return np.nan

df['age'] = df['age'].apply(clean_age)

df['age']
```

```python
#check the values of salary
df['salary'].head
```

```python
#cleaning the values and convert it from string to float
df['salary'] =(
    df['salary']
    .astype(str)
    .replace(r"[$,]", "", regex=True)
    .replace("NaN", np.nan)
    .astype(float)
)

df["salary"]
```

```python
df['performance'] = (
    df['performance']
    .astype(str)
    .replace(r"[%]", "", regex=True)
    .replace("None", np.nan)
    .astype(float)
)
    
df['performance']
```

```python
#clean the inconsistency value in remote
df.replace({
    "yes": "Yes",
    "no": "No",
    "N": "No",
    "Y": "Yes",
    "None": np.nan,
    "True": "Yes"
}, inplace=True)

df
```

```python
#clean inconsistence value in join_date colomn
df['join_date'].to_list()
```

```python
#df['join_date'] = pd.to_datetime(df['join_date'], errors="coerce")
#df['join_date']

def date_parser(date_str):
    try:
        return parser.parse(date_str, dayfirst=True)
    except Exception:
        return pd. NaT

df['join_date'] = df['join_date'].apply(date_parser)
df['join_date']
                                        
```

```python
#clean inconsistence values in employee id
df['emp_id'] = pd.to_numeric(df['emp_id'], errors='coerce')
df['emp_id']
```

```python
df.info()
```

```python
#handling missing values
df['full_name'] = df['full_name'].fillna("Unknown")
df['full_name']
                             
```

```python
df['department'] = df['department'].fillna(df['department'].mode()[0])
df['department']
```

```python
df['age'] = df['age'].fillna(df['age'].median())
df['age']
```

```python
df['salary'] = df['salary'].fillna(df['salary'].median())
df['salary']
```

```python
df['performance'] = df['performance'].fillna(df['performance'].median())
df['performance']
```

```python
df['remote'] = df['remote'].fillna(df['remote'].mode()[0])
df['remote']
```
